```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import glob
import random
```

```python
path = ''
```

```python
NUM_CLASSES = len(glob.glob(path + 'train/*'))
print(NUM_CLASSES)
```

```python
imgs = []
counts = []
labels = []

for folder_path in glob.glob(path + 'train/*'):
    labels.append(folder_path.split('\\')[-1])
    counts.append(len(glob.glob(folder_path + '/*.*')))
    img_path =  random.choice(glob.glob(folder_path + '/*.*'))
    imgs.append(cv2.imread(img_path))
```

```python
plt.figure()

fig, axes = plt.subplots(5, 6, figsize = (16, 16))

for i, img in enumerate(imgs):

    plt.subplot(5, 6, i + 1)
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'{labels[i]}: {counts[i]}')

fig.delaxes(axes.flatten()[NUM_CLASSES])

plt.show()
```
